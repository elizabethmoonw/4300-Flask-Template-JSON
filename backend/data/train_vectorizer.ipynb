```python
import pandas as pd
from preprocess import extract_keywords, best_tags
from constants import MAKEUP_ATTRIBUTES, MAC_REVS, DESC_ONLY_ATTRIBUTES
```

```python
data = pd.read_json('../scraping/all_data.json')
```

```python
products_df = pd.json_normalize(data['products'], 
                       meta=['id', 'product', 'category', 'link_name', 'link', 'brand', 'price', 'img_link', 'ingredient', 'avg_rating', 'reviews', 'summary'],
                       errors='ignore')
products_df
```

```python

print(best_tags(MAKEUP_ATTRIBUTES, [extract_keywords(rev) for rev in MAC_REVS]))
```

```python
tag_list = []
```

```python
for i in range(0, len(products_df)):
  prod_0 = products_df.iloc[i]
  rev_list = list(prod_0['reviews'])
  rev_list.append(prod_0['summary'])
  print(f"product {i}: {prod_0['product']}")
  print(extract_keywords(rev_list[-1]))
  best_tags_list = best_tags(MAKEUP_ATTRIBUTES, [extract_keywords(rev) for rev in rev_list])
  tag_list.append(list(best_tags_list))
  print(best_tags_list)
  print()
```

```python
tag_list
```

```python
products_df['tags'] = tag_list
```

```python
products_df.to_csv('tagged_products.csv')
```
